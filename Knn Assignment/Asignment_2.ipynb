```python
import pandas as pd
from matplotlib import pyplot as plt
```

1. Import data set

```python
df1=pd.read_excel("dataset_2.xlsx")
df1.head()
```

2. Separate X (Gender, Height) and Y (y=Weight).

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Gender'] = le.fit_transform(df1['Gender'])
x=df1[['Gender','Height']]
y=df1[['Weight']]
```

3. Train = 70%, Test = 30%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, train_size=0.7,random_state=1)
```

Apply Linear Regression

```python
from sklearn.linear_model import LinearRegression
LinearRegression=LinearRegression()
LinearRegression.fit(X_train,Y_train)
```

```python
Y_train_predict=LinearRegression.predict(X_train)
print(Y_train_predict)
Y_test_predict=LinearRegression.predict(X_test)
print(Y_test_predict)
```

5. Evaluate the Model (Testing and training Accuracy, MSE for testing)

```python
plt.plot(Y_train,Y_train_predict)
plt.plot(Y_test,Y_test_predict)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
train_sceore=r2_score(Y_train,Y_train_predict)
print(train_sceore)
test_sceore=r2_score(Y_test,Y_test_predict)
print(test_sceore)
test_mse=mean_squared_error(Y_test,Y_test_predict)
print(test_mse)
```

6. Apply KNN Regressor:

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(X_train,Y_train)
```

```python
Y_train_predict_knn = neigh.predict(X_train)
print(Y_train_predict_knn)
Y_test_predict_knn = neigh.predict(X_test)
print(Y_test_predict_knn)
```

7. Evaluate the Model (Testing and training Accuracy, MSE for testing)

```python
knn_train_score = r2_score(Y_train,Y_train_predict_knn)
print(knn_train_score) 
knn_test_score = r2_score(Y_test,Y_test_predict_knn)
print(knn_test_score)
knn_test_mse=mean_squared_error(Y_test,Y_test_predict_knn)
print(knn_test_mse)
```
